```python
key_cols={
    'account':[sales_pipeline,accounts],
    'product':[sales_pipeline,products],
    'sales_agent':[sales_pipeline,sales_teams]
}
for key, dfs in key_cols.items():
    for df in dfs:
        # Check if the column exists before trying to clean it
        if key in df.columns:
            df[key] = df[key].str.strip().str.lower()
```

```python
# Correct the product name in the products dataframe
products['product'] = products['product'].replace('gtx pro', 'gtxpro')
```

```python
#Clean the join key columns in all tables
sales_pipeline.dropna(subset=['account'], inplace=True)
closed_stages = ['Won', 'Lost']
open_stages = ['Engaging', 'Prospecting']

#Split the pipeline into 'open' and 'closed' deals
closed_deals=sales_pipeline[sales_pipeline['deal_stage'].isin(closed_stages)].copy()
open_deals = sales_pipeline[sales_pipeline['deal_stage'].isin(open_stages)].copy()

#Remove duplicates from lookup tables ---
accounts.drop_duplicates(subset=['account'], inplace=True)
products.drop_duplicates(subset=['product'], inplace=True)
sales_teams.drop_duplicates(subset=['sales_agent'], inplace=True)

#JOIN TABLES FOR CLOSED DEALS
master_closed=pd.merge(closed_deals,accounts, on='account', how='left')
master_closed=master_closed.merge(products, on='product', how='left')
master_closed=master_closed.merge(sales_teams, on='sales_agent', how='left')
master_closed.to_csv('master_closed.csv', index=False)

#join tables for open deals
master_open=pd.merge(open_deals,accounts, on='account', how='left')
master_open=master_open.merge(products, on='product', how='left')
master_open=master_open.merge(sales_teams, on='sales_agent', how='left')
master_open.to_csv('master_open.csv', index=False)
```

```python
#FEATURE ENGINEERING (Creating new columns for analysis)
# Convert date columns to datetime objects
master_closed['engage_date'] = pd.to_datetime(master_closed['engage_date'])
master_closed['close_date'] = pd.to_datetime(master_closed['close_date'])

# Calculate sales cycle in days
master_closed['sales_cycle_days'] = (master_closed['close_date'] - master_closed['engage_date']).dt.days

# Discount/Markup column (net difference from list price)
master_closed['discount_markup'] = master_closed['close_value'] - master_closed['sales_price']
```

```python
# Win flag (for ML, optional but useful)
master_closed['win_flag'] = master_closed['deal_stage'].apply(lambda x: 1 if x == 'Won' else 0)
```

```python
# Create a dataframe with only 'Won' deals
won = master_closed[master_closed['deal_stage'] == 'Won'].copy()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Assume 'master_closed' is your loaded DataFrame
# Create a snapshot date for a consistent "today" for the analysis
# This is set to one day after the most recent transaction in your data
snapshot_date = won['close_date'].max() + pd.Timedelta(days=1)
# --- 3. Calculate R, F, and M values for each customer ---
rfm_df = won.groupby('account').agg(
    # Use the snapshot_date for a relative recency calculation
    Recency=('close_date', lambda date: (snapshot_date - date.max()).days),
    Frequency=('opportunity_id', 'count'),
    MonetaryValue=('close_value', 'sum')
).reset_index()

# The rest of your code remains the same
rfm_data = rfm_df[['Recency', 'Frequency', 'MonetaryValue']]

# Scale the data
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_data)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Find the optimal number of clusters
inertia = []
# Limit k to the number of samples
max_k = rfm_scaled.shape[0]
# Check if there are enough samples for clustering
if max_k == 0:
    print("Not enough data to perform clustering.")
else:
    for k in range(1, min(max_k, 11)): # Limit k to a maximum of 10 or the number of samples
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(rfm_scaled)
        inertia.append(kmeans.inertia_)

    # Plot the Elbow Curve
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, min(max_k, 11)), inertia, marker='o', linestyle='--')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.title('Elbow Method for Optimal k')
    plt.grid(True)
    plt.show()
```

```python
# Run K-Means with the optimal k (e.g., 4)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
rfm_df['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
# Analyze the characteristics of each cluster
cluster_analysis = rfm_df.groupby('Cluster')[['Recency', 'Frequency', 'MonetaryValue']].mean().round(0)
print(cluster_analysis)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 9))
sns.scatterplot(
    data=rfm_df,
    x='Recency',
    y='Frequency',
    size='MonetaryValue',
    hue='Cluster', # Color by the new K-Means cluster
    palette='deep', # Use a different color palette
    sizes=(50, 1500),
    alpha=0.8
)
plt.title('Data-Driven K-Means Clusters', fontsize=16)
plt.xlabel('Recency (Days)')
plt.ylabel('Frequency')
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().invert_xaxis()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from datetime import timedelta

# --- 1. Load and Prepare "master_closed.csv" ---

print("Loading data...")
df_closed = pd.read_csv('master_closed.csv')

# Convert date columns
df_closed['engage_date'] = pd.to_datetime(df_closed['engage_date'])
df_closed['close_date'] = pd.to_datetime(df_closed['close_date'])

# Handle missing values
df_closed_clean = df_closed.dropna(subset=['series', 'sales_price']).copy()

# Calculate sales cycle
df_closed_clean['sales_cycle_days'] = (df_closed_clean['close_date'] - df_closed_clean['engage_date']).dt.days

# Filter for only 'Won' deals
df_closed_won = df_closed_clean[df_closed_clean['deal_stage'] == 'Won'].copy()

# Select features and target
features = [
    'sales_price', 'revenue', 'employees', 'year_established',
    'sector', 'series', 'regional_office', 'manager'
]
X = df_closed_won[features]
y = df_closed_won['sales_cycle_days']

# One-hot encode categorical features
X_encoded = pd.get_dummies(X, drop_first=True)


# --- 2. Split Data, Train, and EVALUATE the Model ---

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
print(f"Training on {len(X_train)} deals, testing on {len(X_test)} deals.")

# Initialize and train the model on the training data
model = GradientBoostingRegressor(random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set to evaluate performance
y_pred_test = model.predict(X_test)

# Calculate and print accuracy metrics
mae = mean_absolute_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)

print("\n--- Model Evaluation ---")
print(f"Mean Absolute Error (MAE): {mae:.2f} days")
print(f"R-squared (R²): {r2:.2f}")
print(f"Interpretation: The model's predictions are off by an average of {mae:.0f} days.")


# --- 3. Retrain Model on ALL Data for Final Prediction ---

print("\nRetraining model on all available 'Won' deal data...")
# It's best practice to use all data for the final forecast
model.fit(X_encoded, y)
print("Model training complete.")


# --- 4. Load and Predict on "master_open.csv" ---

print("\nLoading 'master_open.csv' for prediction...")
df_open = pd.read_csv('master_open.csv')

# Prepare open deals data
df_open['engage_date'] = pd.to_datetime(df_open['engage_date'])
df_open_clean = df_open.dropna(subset=['series', 'sales_price']).copy()
print(f"Predicting for {len(df_open_clean)} open deals.")

X_open = df_open_clean[features]
X_open_encoded = pd.get_dummies(X_open, drop_first=True)
X_open_aligned = X_open_encoded.reindex(columns=X_encoded.columns, fill_value=0)

# Make final predictions
df_open_clean['predicted_cycle_days'] = model.predict(X_open_aligned).round(0).astype(int)
df_open_clean['predicted_close_date'] = df_open_clean.apply(
    lambda row: row['engage_date'] + timedelta(days=row['predicted_cycle_days']),
    axis=1
)


# --- 5. Display and Save Results ---

print("\n--- Predictions for Open Deals ---")
final_predictions = df_open_clean[['opportunity_id', 'engage_date', 'sales_price', 'sector', 'predicted_cycle_days', 'predicted_close_date']]
print(final_predictions.head())

output_filename = 'open_deals_forecast_evaluated.csv'
final_predictions.to_csv(output_filename, index=False)
print(f"\nPredictions saved to '{output_filename}'")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. Load and Prepare "master_closed.csv" for Training ---

print("Loading and preparing 'master_closed.csv' for training...")
df_closed = pd.read_csv('master_closed.csv')

# Handle missing values (dropping for simplicity)
df_closed_clean = df_closed.dropna(subset=['series', 'sales_price']).copy()

# Calculate win_flag on the cleaned data
df_closed_clean['win_flag'] = df_closed_clean['deal_stage'].apply(lambda x: 1 if x == 'Won' else 0)


# Select features that are also available in the master_open dataset
features = [
    'sales_price', 'revenue', 'employees', 'year_established',
    'sector', 'series', 'regional_office', 'manager'
]
X = df_closed_clean[features]
y = df_closed_clean['win_flag'] # Our target is now win_flag

# One-hot encode categorical features
X_encoded = pd.get_dummies(X, drop_first=True)

# Split data for training and evaluation
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42, stratify=y)
print(f"Training model on {len(X_train)} deals and testing on {len(X_test)} deals.")


# --- 2. Train the Classification Model ---

print("Training classification model...")
# Using GradientBoostingClassifier for this task
classifier = GradientBoostingClassifier(random_state=42)
classifier.fit(X_train, y_train)
print("Model training complete.")


# --- 3. Evaluate the Model's Performance ---

y_pred_test = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred_test)
print(f"\nModel Accuracy on Test Data: {accuracy:.2%}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred_test))

# Visualize the Confusion Matrix
cm = confusion_matrix(y_test, y_pred_test)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
plt.title('Confusion Matrix')
plt.ylabel('Actual Outcome')
plt.xlabel('Predicted Outcome')
plt.show()


# --- 4. Predict on "master_open.csv" ---

print("\nLoading 'master_open.csv' for prediction...")
df_open = pd.read_csv('master_open.csv')
df_open_clean = df_open.dropna(subset=['series', 'sales_price']).copy()

# Prepare the prediction data exactly like the training data
X_open = df_open_clean[features]
X_open_encoded = pd.get_dummies(X_open, drop_first=True)

# Align columns to ensure the prediction set matches the training set
X_train_cols = X_train.columns
X_open_aligned = X_open_encoded.reindex(columns=X_train_cols, fill_value=0)

# Predict both the probability AND the final outcome

# Predict the probability of winning (class 1)
win_probabilities = classifier.predict_proba(X_open_aligned)[:, 1]
df_open_clean['win_probability'] = win_probabilities

# Predict the direct Win/Loss outcome (1 or 0)
win_loss_prediction = classifier.predict(X_open_aligned)
# Create a readable label
df_open_clean['predicted_outcome'] = np.where(win_loss_prediction == 1, 'Win', 'Loss')


# --- 5. Display and Save Results ---

print("\n--- Win/Loss Forecast for Open Deals ---")
# Add the new column to the final output
final_predictions = df_open_clean[['opportunity_id', 'sales_price', 'sector', 'win_probability', 'predicted_outcome']]
print(final_predictions.sort_values(by='win_probability', ascending=False).head(10))

# Save the results to a new CSV file
output_filename = 'open_deals_win_loss_forecast.csv'
final_predictions.to_csv(output_filename, index=False)
print(f"\nPredictions saved to '{output_filename}'")
```
